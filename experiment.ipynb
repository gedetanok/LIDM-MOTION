```python
# Convert the uploaded JSON exercises into a CSV suitable for Supabase import
import json, pandas as pd, os

src_json = "exercises_geometry_seed.json"
dst_csv = "exercises_geometry_seed.csv"

# Load JSON
with open(src_json, "r", encoding="utf-8") as f:
    data = json.load(f)

# Keep only the columns that exist on the Supabase table
rows = [
    {
        "title": item.get("title", ""),
        "question_text": item.get("question_text", ""),
        "answer": item.get("answer", ""),
        "difficulty": item.get("difficulty", ""),
    }
    for item in data
]

df = pd.DataFrame(rows, columns=["title", "question_text", "answer", "difficulty"])

# Save to CSV without index

```

```python
df.to_csv(dst_csv, index=False)
```
